```python
import numpy as np
X = np.array([[0,0],
              [0,1],
              [1,0],
              [1,1]])

y = np.array([[0],
              [1],
              [1],
              [0]])
np.random.seed(1)
W1 = np.random.randn(2, 4) 
b1 = np.zeros((1, 4))
W2 = np.random.randn(4, 1)   
b2 = np.zeros((1, 1))
lr = 0.1
for epoch in range(1000):
    z1 = X @ W1 + b1
    a1 = np.maximum(0, z1)       
    z2 = a1 @ W2 + b2
    output = z2                 
    loss = np.mean((output - y) ** 2)
    d_output = output - y
    dW2 = a1.T @ d_output
    db2 = np.sum(d_output, axis=0, keepdims=True)
    da1 = d_output @ W2.T
    dz1 = da1 * (z1 > 0)
    dW1 = X.T @ dz1
    db1 = np.sum(dz1, axis=0, keepdims=True)
    W2 -= lr * dW2
    b2 -= lr * db2
    W1 -= lr * dW1
    b1 -= lr * db1
    if epoch % 200 == 0:
        print(f"Epoch {epoch}, Loss: {loss:.4f}")
print("\nPredictions:")
print(output)
```
