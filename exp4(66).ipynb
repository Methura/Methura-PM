```python
import numpy as np
x = np.array([[0,0],
              [0,1],
              [1,0],
              [1,1]])
y = np.array([[1,0],   
              [0,1],
              [0,1],
              [1,0]])
np.random.seed(1)
w1 = np.random.randn(2,8)
b1 = np.zeros((1,8))
w2 = np.random.randn(8,6)
b2 = np.zeros((1,6))
w3 = np.random.randn(6,4)
b3 = np.zeros((1,4))
w4 = np.random.randn(4,2)
b4 = np.zeros((1,2))
lr = 0.66
epochs = 500
def sigmoid(z): return 1/(1+np.exp(-z))
def dsigmoid(a): return a*(1-a)
def tanh(z): return np.tanh(z)
def dtanh(a): return 1-a**2
def relu(z): return np.maximum(0,z)
def drelu(z): return (z>0).astype(float)
def softmax(z):
    exp_z = np.exp(z - np.max(z,axis=1,keepdims=True))
    return exp_z / np.sum(exp_z,axis=1,keepdims=True)
def swish(z): return z*sigmoid(z)
def dswish(z): return swish(z) + sigmoid(z)*(1-swish(z))
def train(activation_hidden, activation_output, dactivation_hidden):
    global w1,w2,w3,w4,b1,b2,b3,b4
    w1 = np.random.randn(2,8); b1 = np.zeros((1,8))
    w2 = np.random.randn(8,6); b2 = np.zeros((1,6))
    w3 = np.random.randn(6,4); b3 = np.zeros((1,4))
    w4 = np.random.randn(4,2); b4 = np.zeros((1,2))
    for epoch in range(epochs):
        z1 = x@w1 + b1; a1 = activation_hidden(z1)
        z2 = a1@w2 + b2; a2 = activation_hidden(z2)
        z3 = a2@w3 + b3; a3 = activation_hidden(z3)
        z4 = a3@w4 + b4; output = activation_output(z4)
        loss = np.mean((output-y)**2)
        d_output = (output-y)
        dw4 = a3.T @ d_output
        db4 = np.sum(d_output,axis=0,keepdims=True)
        da3 = d_output @ w4.T
        dz3 = da3 * dactivation_hidden(a3)
        dw3 = a2.T @ dz3
        db3 = np.sum(dz3,axis=0,keepdims=True)
        da2 = dz3 @ w3.T
        dz2 = da2 * dactivation_hidden(a2)
        dw2 = a1.T @ dz2
        db2 = np.sum(dz2,axis=0,keepdims=True)
        da1 = dz2 @ w2.T
        dz1 = da1 * dactivation_hidden(a1)
        dw1 = x.T @ dz1
        db1 = np.sum(dz1,axis=0,keepdims=True)
        w4 -= lr*dw4; b4 -= lr*db4
        w3 -= lr*dw3; b3 -= lr*db3
        w2 -= lr*dw2; b2 -= lr*db2
        w1 -= lr*dw1; b1 -= lr*db1
        if epoch % 100 == 0:
            print(f"Epoch {epoch}, Loss: {loss:.4f}")
            print("Predictions:\n", output)
print("Using Sigmoid")
train(sigmoid, sigmoid, dsigmoid)

print("Using Tanh")
train(tanh, sigmoid, dtanh)

print("Using ReLU")
train(relu, sigmoid, drelu)

print("Using Swish")
train(swish, sigmoid, dswish)

print("Using Softmax")
train(relu, softmax, drelu)
```
