```python
import numpy as np 
 

def step_function(net): 
    return 1 if net >= 0 else 0 
 

def train_perceptron(X, y, w, b, lr):  
    epoch = 1 
    while True: 
        error_count = 0 
        print('Epoch:',epoch) 
        for i in range(len(X)): 
            net = np.dot(X[i], w) - b 
            output = step_function(net) 
            error = y[i] - output 
             
            print("Input:",X[i]) 
            print("Net:",net) 
            print("Output:",output) 
            print("Target:",y[i]) 
             
            if error != 0: 
                w = w + lr * error * X[i] 
                b = b-lr * error 
                error_count += 1 
                print("Updated weights:",w)
                print("Updated bias:",b) 
         
        if error_count == 0: 
            break 
        epoch += 1 
     
    print("Final weights:",w) 
    print("Final bias:",b) 
    return w, b 
 
# Input patterns 
X = np.array([ 
    [0, 0], 
    [0, 1], 
    [1, 0], 
    [1, 1] 
]) 
 
# ---------------- OR Gate ---------------- 
y_or = np.array([0, 1, 1, 1]) 
w_or = np.array([0.2, 0.2]) 
b_or = -0.1 
lr = 0.1 
 
w_or, b_or = train_perceptron(X, y_or, w_or, b_or, lr) 
# ---------------- AND Gate ---------------- 
y_and = np.array([0, 0, 0, 1]) 
w_and = np.array([0.2, 0.2]) 
b_and = -0.3 
w_and, b_and = train_perceptron(X, y_and, w_and, b_and, lr) 
```
